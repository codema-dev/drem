```python
import pandas as pd
```

```python
%store -r BER_Dublin

BER_Dublin.head()
```

```python
BER_overall_Dublin_archetypes = pd.DataFrame()

BER_overall_Dublin_archetypes = BER_Dublin.pivot_table(
    index=["Dwelling Type", "Period Built"],
    aggfunc={
        "Postcodes": "count",
        "DeliveredEnergyMainSpace (ind)": "mean",
        "DeliveredEnergyMainWater (ind)": "mean",
    },
    dropna=False,
)

BER_overall_Dublin_archetypes.rename(columns={
    "Postcodes":"Number of Samples",
    "DeliveredEnergyMainSpace (ind)":"Average BER SH Demand [MWh/year]",
    "DeliveredEnergyMainWater (ind)":"Average BER HW Demand [MWh/year]",
},inplace=True)

# Fill all rows where no sample was found with 0
BER_overall_Dublin_archetypes["Number of Samples"].fillna(0, inplace=True)

# print("Postcode-specific archetypes")
BER_overall_Dublin_archetypes
```
