```python
import pandas as pd
```

```python
%store -r BER_Dublin

BER_Dublin.head()
```

```python
BER_Dublin_archetypes = pd.DataFrame()

BER_Dublin_archetypes = BER_Dublin.pivot_table(
    index=["Postcodes", "Dwelling Type", "Period Built"],
    aggfunc={
        "Postcodes": "count",
        "DeliveredEnergyMainSpace (ind)": "mean",
        "DeliveredEnergyMainWater (ind)": "mean",
    },
    dropna=False,
)

BER_Dublin_archetypes.rename(columns={
    "Postcodes":"Number of Samples",
    "DeliveredEnergyMainSpace (ind)":"Average BER SH Demand [MWh/year]",
    "DeliveredEnergyMainWater (ind)":"Average BER HW Demand [MWh/year]",
},inplace=True)

# Fill all rows where no sample was found with 0
BER_Dublin_archetypes["Number of Samples"].fillna(0, inplace=True)

# print("Postcode-specific archetypes")
BER_Dublin_archetypes.head()
```

```python
## BUG IN PANDAS - original Postcodes of BERPublicsearch reappear with pivot_table operation

BER_Dublin_archetypes.reset_index(inplace=True)

# Use mask to access all "Dublin" data rows
mask = BER_Dublin_archetypes["Postcodes"].str.contains("Dublin")
BER_Dublin_archetypes = BER_Dublin_archetypes.loc[mask].copy(deep=True)

BER_Dublin_archetypes.set_index(
    ["Postcodes", "Dwelling Type", "Period Built"], inplace=True
)

BER_Dublin_archetypes.head()
```
