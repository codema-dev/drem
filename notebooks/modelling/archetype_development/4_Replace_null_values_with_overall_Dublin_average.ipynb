```python
import pandas as pd
```

```python
%store -r BER_Dublin_archetypes
%store -r BER_overall_Dublin_archetypes
```

```python
BER_Dublin_archetypes["Postcode-specific data?"] = True

print("Dublin 1 before replacement:")
idx = pd.IndexSlice
BER_Dublin_archetypes.loc[idx["Dublin 1",:,:]].head(15)
```

```python
## Fill all rows with < 100 samples with overall Dublin averages


# Set "Postcode-specific data?" to False for rows with < 100 samples
mask = BER_Dublin_archetypes["Number of Samples"] < 100
BER_Dublin_archetypes.loc[mask, "Postcode-specific data?"] = False

# Fill empty values with Dublin data
BER_Dublin_archetypes = BER_Dublin_archetypes.combine_first(BER_overall_Dublin_archetypes)
```

```python
## Reorder df in original form


def pull_numbers_from_postcodes(x):

    if x == "Co. Dublin":
        return -1
    elif x == "Dublin 6W":
        x = x.split()[1]  # Get 6W only where [0,1]==['Dublin','6W']
        return int(x[0]) + 0.1  # Return 6 only
    else:
        x = x.split()
        return int(x[1])


BER_Dublin_archetypes.reset_index(inplace=True)
BER_Dublin_archetypes["Postcode Numbers"] = BER_Dublin_archetypes["Postcodes"].apply(
    pull_numbers_from_postcodes
)
BER_Dublin_archetypes.sort_values(
    ["Postcode Numbers", "Dwelling Type", "Period Built"], inplace=True
)
del BER_Dublin_archetypes["Postcode Numbers"]  # Delete number column
BER_Dublin_archetypes.set_index(
    ["Postcodes", "Dwelling Type", "Period Built"], inplace=True
)

print("Dublin 1 after replacement:")
BER_Dublin_archetypes.loc[idx["Dublin 1",:,:]].head(15)

# Source: https://stackoverflow.com/questions/39816671/pandas-populate-new-dataframe-column-based-on-matching-columns-in-another-datafr
```
