```python
def calculate_sample_statistics(df):

    print(df)
    
    sample = df.sample(n=100, random_state=1)

    sample_statistics = pd.DataFrame()
    sample_statistics["Postcodes"] = df.name

    sample_statistics[["HW_mean", "SH_mean"]] = sample[
        ["DeliveredEnergyMainWater (ind)", "DeliveredEnergyMainSpace (ind)"]
    ].mean()
    
    sample_statistics[["HW_sem", "SH_sem"]] = sample[
        ["DeliveredEnergyMainWater (ind)", "DeliveredEnergyMainSpace (ind)"]
    ].sem()  # = Standard Error of Mean

    print(sample_statistics)

calculate_sample_statistics = np.vectorize(calculate_sample_statistics)

# def get_sample_statistics(df):

#     print(df.shape[0])

#     statistics = statistics.append(calculate_sample_statistics(df))


#     return
```

```python
BER_Dublin_postcode_estimates = BER_Dublin_postcodes.agg(calculate_sample_statistics)
```
