```python
def convert_year_construction_to_CSO_period_built(year_built):

    if year_built == None:
        return "Not stated*"
    elif year_built < 1920:
        return "before 1919*"
    elif year_built < 1946:
        return "1919 - 1945*"
    elif year_built < 1961:
        return "1946 - 1960*"
    elif year_built < 1971:
        return "1961 - 1970*"
    elif year_built < 1981:
        return "1971 - 1980*"
    elif year_built < 1991:
        return "1981 - 1990*"
    elif year_built < 2001:
        return "1991 - 2000*"
    elif year_built < 2006:
        return "2001 - 2005*"
    elif year_built < 2020:
        return "2006 or later*"
    else:
        print(year_built)
        return None


convert_year_construction_to_CSO_period_built = np.vectorize(
    convert_year_construction_to_CSO_period_built
)

BER_Dublin_specific[
    "Period Built"
] = convert_year_construction_to_CSO_period_built(
    BER_Dublin_specific["Year_of_Construction"].values
)

BER_Dublin_specific["Period Built"] = BER_Dublin_specific[
    "Period Built"
].astype("category")

BER_Dublin_specific[["Postcodes","Period Built"]].head()
```
